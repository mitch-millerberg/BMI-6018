## Type Conversion

Type conversion means a variable originally declared as a particular data type is changed (cast) to another

implicit vs. explicit (typecasting)

```python
#initialize variables
x = 'hello'
y = 0
```

```python
#show type of variable
print(type(x))
print(type(y))    
```

```python
#print as bool values
#explicit type conversion (typecasting)
print(bool(x))
print(bool(y))
```

```python
#print type of x,y after being cast as bools
print(type(bool(x)))
print(type(bool(y)))
```

```python
print(x)
print(type(x))
```

```python
#can cast and save in original variable name
x = bool(x)
print(x)
print(type(x))
```

```python
#empty string
print(bool(''))
#keyword None (null value)
print(bool(None))
```

## Boolean Operators and If conditions...

| Boolean | True, True | True, False | False, True | False, False |
| ------- | ---------- | ----------- | ----------- | ------------ |
| And (&) | True       | False       | False       | False        |
| Or (\|) | True       | True        | True        | False        |
| XOR (^) | False      | True        | True        | False        |
| NAND    | False      | True        | True        | True         |
| NOR     | False      | False       | False       | True         |
| XNOR    | True       | False       | False       | True         |

```python
mask = True
hand_wash = False

#True and False
if (mask == True) and (hand_wash == True):
    print('You are safe!')

#False or True
if (mask == False) or (hand_wash == False):
    print('You need to be Covid tested!')
```
